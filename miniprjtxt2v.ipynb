```python
# Install necessary packages
!pip install modelscope==1.4.2
!pip install open_clip_torch==2.24.0
!pip install pytorch-lightning

from modelscope.pipelines import pipeline
from modelscope.outputs import OutputKeys
import moviepy.editor as mp

# Set up pipeline
p = pipeline('text-to-video-synthesis', 'damo/text-to-video-synthesis')

# Define prompt and output details
test_text = {
    'text': 'Create a high-definition, clear video of a playful cat interacting with various objects over 30 seconds. The sequence starts with the cat jumping over colorful obstacles, then transitioning to chasing a vibrant, rolling ball. After playing, the cat pauses to groom itself by a sunny window, capturing soft light and shadows. Finally, the cat curls up to sleep peacefully, with the background styled in swirling, vivid colors inspired by Van Gogh’s brush strokes. Each action should transition smoothly with consistent, high-quality visuals and fine details in a minimum of 30 seconds.',
}

# Generate a short video (the model may not support 30-second video generation directly)
output_clip_path = p(test_text, output_video='./output.mp4')[OutputKeys.OUTPUT_VIDEO]

# Repeat the clip to reach approximately 30 seconds
clip = mp.VideoFileClip(output_clip_path)
duration = clip.duration

# Number of times to loop to approximate 30 seconds
n_loops = int(30 / duration) + 1
final_clip = mp.concatenate_videoclips([clip] * n_loops).subclip(0, 30)

# Save the final extended video
final_output_path = './output_30sec.mp4'
final_clip.write_videofile(final_output_path, codec="libx264")

print('Extended output video path:', final_output_path)
```
